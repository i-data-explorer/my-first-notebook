## 2. How to use a Notebook in Jupyterlab 3.0?

<a id="2.1"></a>
### 2.1. Cells - Basic Block of a Notebook

A Jupyter notebook is designed for exploration, and consists of blocks of python code known as ```Cell```. These cells can be independently executed, with content accessible across cell. 

Cells can contain the following
1. Code
2. Markdown
3. Raw NBConvert
4. Heading

<a id="2.2"></a>
### 2.2. Adding a Cell

To add a new `Cell`, click the `+` button in the shortcut bar as shown.
    


<a id="2.3"></a>
### 2.3. Running a Cell

To run a Cell, it must be selected, by simply clicking on the ``Cell`` and then can be executed by clicking the ``Run`` symbol on the shortcut bar, by using the keyboard shortcut ``Ctrl + Enter`` or the ``Run`` option in the Menu bar.

A Cell is selected if it has a blue vertical bar to the left of it, as shown in the image below.



````python
# This is an sample cell.
# Select this cell and click the "Run" icon on the top bar, or press Ctrl+Enter to run it.

def greet_world(username):
    print("Hello", username, ", welcome to Jurassic Park of Automated Python programmed AI Dinosaurs")
    
greet_world("Jupyter-user")
````

<a id="2.4"></a>
### 2.4. Running multiple Cells and more...

Multiple cells can be run from the ``Run`` option in the Menu bar. Additionally, there are several other run options available in the dropdown menu.



<a id="2.5"></a>
### 2.5. Editing Cells

A cell can be copied, cut, pasted, moved, merged, split and deleted using the ``Edit`` option. The associated keyboard shortcuts are also mentioned, and makes it convenient.



<a id="2.6"></a>
### 2.6. File Options

The `File` option in the menu bar, has a variety of essential options such as `Download`, `Save`, `Export` or `Rename notebook`. 



<a id="2.7"></a>
### 2.7. Kernel - The Computation Engine

A notebook kernel is a computational engine that executes the code contained in a Notebook document. Sometimes, there is a requirement for the kernel to be restarted in cases of failure, or stopped for unwanted executions. The kernel options are made available in `Kernel`in the menu bar.



<a id="2.8"></a>
### 2.8. Rename Notebook

A notebook can be renamed by right clicking on the `notebook tab` and selecting `Rename Notebook...` option.
It can also be renamed by visitng the `File` option in the menu bar, and selecting the `Rename Notebook...` option within it.



<a id="2.9"></a>
### 2.9 Save Notebook

The notebook can either be saved by clicking the `Save symbol` in the shortcut bar, or from the `File` option in the menu bar or with a simple keyboard shortcut, `Ctrl+S`.



<a id="2.10"></a>
### 2.10. Handy Shortcuts

Some shortcuts which are often used include the following.

1. `Ctrl+Enter` to execute cell(s)
2. `d d` to delete a cell
3. Clicking outside the text area of a cell, to the left of it will allow us to select a cell and drag it up or down, shifting positions.
4. `Ctrl+S` to save the notebook

Additional Shortcuts are present in the shortcut bar on the top of the notebook.



Options provided in the shortcut bar include

1. Save
2. Add cell
3. Cut cell
4. Copy cell
5. Paste cell
6. Execute cell
7. Stop execution
8. Restart Kernel
9. Restart and run all cells
10. Set type of cell
